```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
image = np.random.randint(0,256,(100,100,3),dtype=np.uint8)

plt.imshow(image)
plt.title('Random RGB image')
plt.axis('off')
plt.show()
```

```python
from skimage.transform import rotate

rotated = rotate(image,20,mode='edge')
print('Original image shape',image.shape)
print('Rotated image shape',rotated.shape)
```

```python
plt.figure(figsize=(4,4))
plt.subplot(1,2,1)
plt.imshow(image)
plt.title('Original')
plt.axis('off')


plt.subplot(1, 2, 2)
plt.imshow(rotated)
plt.title('Rotated (mode=edge)')
plt.axis('off')
```

```python
img = np.random.randint(0,256,(100,100,3),dtype=np.uint8)
img[30:70,30:70] = [255,255,255]
```

```python
modes = ['constant','edge','wrap','symmetric','reflect']

rotated_images = [rotate(img,angle = 30,mode = mode,preserve_range=True).astype(np.uint8) 
                 for mode in modes]
```

```python
plt.figure(figsize=(15,4))

for i,(mode,img) in enumerate((zip(modes,rotated_images))):
    plt.subplot(1,len(rotated_images),i+1)
    plt.imshow(img)
    plt.title(f"mode='{mode}'")
    plt.axis('off')


plt.suptitle("Rotation with Different Modes", fontsize=16)
plt.tight_layout()
plt.show()    
```
